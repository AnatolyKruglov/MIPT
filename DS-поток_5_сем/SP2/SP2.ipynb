## Задача 1. 


**1.** Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $U[0, \theta]$. Рассмотрим оценки $X_{(n)}, \frac{n+1}{n}X_{(n)}, 2\overline{X}$ параметра $\theta$. Какие из этих оценок являются несмещенными?

_**Ответ:**_  
$X_{(n)}$ - _<span style="color: red;">смещенная</span>, но ее матожидание стремится к параметру в пределе при $n \to \infty$, т.е. оценка - асимптотически несмещенная (доказано на семинаре 10.09)_  
$\frac{n+1}{n}X_{(n)}$ - _<span style="color: green;">несмещенная</span> (доказано в дз ST2, задача 1.д)_  
$2\overline{X}$ - _<span style="color: green;">несмещенная</span> (доказано в дз ST2, задача 1.а)_  

Проведите практическое исследование несмещенности перечисленных оценок аналогично ислледованию несмещенности выборочной дисперсии в предыдущих заданиях. Разбор этой задачи для выборочной дисперсии был на занятии.

**Решение:**  

```python
estimators = [
    (lambda X: X.max(axis=-1), "$X_{(n)}$", "red", 0.2),
    (lambda X: (X.shape[1] + 1) / X.shape[1] * X.max(axis=-1), "$\\frac{n+1}{n}X_{(n)}$", "green", 0.1),
    (lambda X: 2 * X.mean(axis=-1), "$2\overline{X}$", "green", 0.0)
]

sample_count = 500 
for sample_size in (10, 100, 500):       
    X = sps.uniform.rvs(size=(sample_count, sample_size))  # кол-во экспериментов * размер выборки
    plt.figure(figsize=(18, 0.7 * len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1, s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300, color='w', edgecolors='black')
    plt.vlines(1, -1, 1, color='r')
    plt.title('Размер выборки = %d' % sample_size)
    plt.yticks([])
    plt.legend()
    plt.xlim((0.7, 1.3))
    plt.ylim((-0.1, 0.3))
```

_**Вывод:**_  
_Оценку называют несмещенной, если ее матожидание равно параметру (что важно: при любом значении параметра). В этой задаче наглядно видно, что <span style="color: green;">несмещенные</span> оценки хорошо работают даже при малом размере выборки, а <span style="color: red;">смещенная</span> (асимптотически несмещенная) - нет (только в пределе с ростом размера выборки). Именно из-за этого свойства несмещенные оценки полезны, например, в физике, где размер выборки зачастую ограничен, а параметр всё равно требуется оценить как можно точнее_

Пусть $X = (X_1, ..., X_n)$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$, причем $a$ не известно. Визуализируйте асимптотический доверительный интервал Вальда для параметра дисперсии при росте размера выборки. 

Для этого сгенерируйте выборку $X_1, ... X_{N}, N = 100$ и постройте график доверительных интервалов уровня доверия $\alpha=0.95$, вычисленных для всех подвыборок размера $n$ вида $X_1, ... X_n$, $1 \le n \le 100$, используя написанную выше функцию. На график также нужно нанести точки выборки.

Чтобы не плодить код, напишите следующую функцию (см. ниже). При выборе стиля графика помните, что если изображаются лишь точки и линии, то лучше использовать серый фон, а если присутствуют закрашенные области, то предпочтительнее белый. Графики первого типа еще называют "легкими", а второго — "тяжелыми".

Для вычисления квантилей у каждого распределения из `scipy.stats` используйте функцию `ppf`.

**Решение:**

<b><font color="#00b565" size="3">Замечания:</font></b> <i>  
▫️ Нет никаких формул для интервалов. Комм. Не выписаны явно формулы для интервалов, они есть только в коде. См. замечание в начале условия задания. (-2.0)  
▫️ ~~Получен неправильный интервал. (-10.0)~~  (интервалы из моего и из официального решения тождественны, по итогу апелляции был засчитан полный балл))
</i>

У меня в решении: $\left(S^2 + z_{(1-\alpha)/2} \cdot \dfrac{\sqrt{2 (S^2)^2}}{\sqrt{n}} , S^2 + z_{(1+\alpha)/2}\cdot \dfrac{\sqrt{2 (S^2)^2}}{\sqrt{n}} \right)$  

<b><font color="#00b565" size="5">===== Конец исправлений =====</font></b>

```python
def draw_confidence_interval(
    left, right, estimation=None, sample=None, ylim=(None, None), 
    color_estimation='#FF3300', color_interval='#00CC66', color_sample='#0066FF', label_estimation='Оценка'
):
    plt.figure(figsize=(14, 8))
    plt.fill_between(range(len(sample)), left, right, color=color_interval, alpha=0.5)  # [i, i] -> range(1, n + 1)
    if estimation is not None:  # оценка
        plt.plot(range(len(sample)), estimation, color=color_estimation, label=label_estimation, marker='o')
    if sample is not None:  # выборка
        plt.scatter(range(len(sample)), sample, color=color_sample, label='Выборка', zorder=5)
    if ylim != (None, None):
        plt.ylim(ylim)
    plt.xlabel('Размер выборки n')
    plt.ylabel('Довер. интервал для дисперсии')
    plt.title('Довер. интервалы для дисперсии vs размер выборки')
    plt.legend()
    plt.grid()
    plt.show()

a, sigma, N = 10, 2, 100
X = np.random.normal(a, sigma, N)
left_bounds, right_bounds, estimations = [], [], []
alpha = 0.95

for n in range(1, N+1):
    sample = X[:n]
    sample_var = np.var(sample, ddof=1) 
    left  = sample_var + norm.ppf((1 - alpha) / 2) * np.sqrt(2 * sample_var**2 / n)  # это тождестенно официальному решению) 
    right = sample_var + norm.ppf((1 + alpha) / 2) * np.sqrt(2 * sample_var**2 / n)
    left_bounds.append(left)
    right_bounds.append(right)
    estimations.append(sample_var)

left_bounds, right_bounds, estimations = np.array(left_bounds), np.array(right_bounds), np.array(estimations)

draw_confidence_interval(
    left=left_bounds,
    right=right_bounds,
    estimation=estimations,
    sample=X,
    ylim=(-10, 20)
)
```

_**Вывод:**  чем больше размер выборки, тем точнее можно оценить параметр (сузить доверительный интервал)_

Реализуйте метод простого градиентного спуска для выше описанной задачи оптимизации. Для каждой итерации сохраните текущие значения среднего и дисперсии, постройте график зависимости значений $a$ и $\sigma^2$ от шага процедуры. Наблюдается ли сходимость к оптимальным параметрам?

```python
def viz(a_values, sigma2_values):
    """Визуализация процесса оптимизации параметров"""
    plt.figure(figsize=(14, 4))

    plt.subplot(1, 2, 1)
    plt.plot(a_values, label='a')
    plt.axhline(y=1, color='black', linestyle='--', label='Опт. знач. a = 1')
    plt.title('Значение a со временем')
    plt.xlabel('Шаг')
    plt.ylabel('a')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(sigma2_values, label='σ²', color='orange')
    plt.axhline(y=1, color='black', linestyle='--', label='Опт. знач. σ² = 1')
    plt.title('Значение σ² со временем')
    plt.xlabel('Шаг')
    plt.ylabel('σ²')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
a, sigma2 = 2, 2  # начальные значения
a_values, sigma2_values = [], []
eta = 0.1  # learning rate

for i in range(1000):  # 1000 итераций
    a_values.append(a)
    sigma2_values.append(sigma2)
    grad_a = 2 * (a - 1)
    grad_sigma2 = 2 * (sigma2 - 1)
    a -= eta * grad_a
    sigma2 -= eta * grad_sigma2

viz(a_values, sigma2_values)
```

А именно, каждый раз мы будем генерировать выборку размера 5 из нормального распределения с текущими значениями параметров и далее по ней оценивать градиент.
Для оценки математического ожидания ипользуйте несмещенную оценку, а для дисперсии &mdash; смещенную. 

Реализуйте описанный выше подход. Как и прежде, изобразите текущие значения параметров в зависимости от итерации. Сошлась ли такая процедура к оптимальным значениям?

```python
a, sigma2 = 2, 2  # начальные значения
a_values, sigma2_values = [], []
eta = 0.1  # learning rate

for i in range(1000):  # 1000 итераций
    a_values.append(a)
    sigma2_values.append(sigma2)
    sample = np.random.normal(loc=a, scale=np.sqrt(sigma2), size=5)
    grad_a = np.mean(sample) - 1  # по a - несмещенная оценка
    grad_sigma2 = np.var(sample, ddof=0) - 1  # по sigma2 - смещенная оценка
    a -= eta * grad_a
    sigma2 -= eta * grad_sigma2

viz(a_values, sigma2_values)
```

*Видно, что такая процедура __не сошлась__ (как минимум, __сошлась сильно хуже__). Значения параметров хоть и приблизились к оптимуму, но все же колебания остаются слишком большими, чтобы считать такой результат приемлемым. Кроме того, видны различия для a и $\sigma^2$: в то время как а колеблется вокруг оптимума (несмещенная оценка), $\sigma^2$ колеблется выше оптимума (смещенная оценка)*

*Мне еще было интересно увеличить размер генерируемой выборки. Гипотетически, это должно сгладить шум и улучшить сходимость*

```python
a, sigma2 = 2, 2  # начальные значения
a_values, sigma2_values = [], []
eta = 0.1  # learning rate

for i in range(1000):  # 1000 итераций
    a_values.append(a)
    sigma2_values.append(sigma2)
    sample = np.random.normal(loc=a, scale=np.sqrt(sigma2), size=500)
    grad_a = np.mean(sample) - 1  # по a - несмещенная оценка
    grad_sigma2 = np.var(sample, ddof=0) - 1  # по sigma2 - смещенная оценка
    a -= eta * grad_a
    sigma2 -= eta * grad_sigma2

viz(a_values, sigma2_values)
```

*Действительно, __в таком обновленном варианте процедура сошлась__ заметно лучше: после 40 шагов графики параметров флуктуируют вблизи оптимального значения. Для практических целей такого качества оценки наверняка было бы достаточно. Однако на практике выборку большого размера получить может быть тяжело, поэтому стоит поискать другую процедуру (использующую несмещенную оценку). Вернемся к заданию*

Теперь изменим нашу процедуру, взяв несмещенную оценку для дисперсии. Обратите внимание на параметр `ddof` в функции `np.var`. Поменялся ли результат?

```python
a, sigma2 = 2, 2  # начальные значения
a_values, sigma2_values = [], []
eta = 0.1  # learning rate

for i in range(1000):  # 1000 итераций
    a_values.append(a)
    sigma2_values.append(sigma2)
    sample = np.random.normal(loc=a, scale=np.sqrt(sigma2), size=5)
    grad_a = np.mean(sample) - 1  # по a - несмещенная оценка
    grad_sigma2 = np.var(sample, ddof=1) - 1  # по sigma2 - тоже несмещенная оценка
    a -= eta * grad_a
    sigma2 -= eta * grad_sigma2

viz(a_values, sigma2_values)
```
