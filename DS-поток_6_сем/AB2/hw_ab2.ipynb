### Задача 3.

Провалидируйте и сравните методы CUPED через линейную регрессию:

На занятии мы рассмотрели несколько методов обобщения CUPED с помощью линейной регрессии:
* стандартный двухшаговый,
* комбинированный одношаговый,
* комбинированный двухшаговый.

Провалидируйте эти методы по величине вероятности ошибки I рода и сравните их по мощности. В качестве данных рассмотрите
* простые искусственные данные из ноутбука с занятия,
* искусственные данные с нелинейной зависимостью от ковариаты и гетероскедастичностью.

```python
def generate_train_test_sample(sample_size=1000, shape=5, scale=1000, factor=0.3, complex=False):
    '''Генерация всей совокупности объектов'''
    base = np.random.gamma(shape=shape, scale=scale, size=sample_size)
    data = pd.DataFrame()
    func = (lambda x: (x**5-np.sin(x)+1)) if complex else (lambda x:x)  # учет более сложной зависимости при complex=True
    data['pilot'] = func(base) + np.random.normal(size=sample_size, scale=base.std()*factor)
    data['prepilot'] = func(base) + np.random.normal(size=sample_size, scale=base.std()*factor)
    return data

def generate_two_independent_samples(sample_size=1000, shape=5, scale=1000, factor=0.3, complex=False):
    '''Генерация двух независимых выборок'''
    return [generate_train_test_sample(sample_size, shape, scale, factor, complex) for _ in range(2)]

def cuped_transform(x_data, y_data, target='pilot', covariate='prepilot', return_theta=False):
    '''CUPED-преобразование'''
    theta = (np.cov(x_data[covariate],x_data[target])[0,1] + np.cov(y_data[covariate],y_data[target])[0,1]) / (x_data[covariate].var(ddof=0) + y_data[covariate].var(ddof=0))
    x_data['cuped'], y_data['cuped'] = x_data[target] - theta*x_data[covariate], y_data[target] - theta*y_data[covariate]
    return x_data, y_data, theta if return_theta else x_data, y_data

def cuped_ttest(x_data, y_data, target='pilot', covariate='prepilot', alpha=0.05):
    '''Абсолютный t-test'''
    x_data, y_data = cuped_transform(x_data, y_data, target=target, covariate=covariate)
    return sps.ttest_ind(x_data['cuped'], y_data['cuped'])

def estimate_reject_prob(n_rejects, n_iter):
    '''Оценка вер-ти отвержения критерия и ее дов. интервала. Используется для оценки вер-ти ошибки первого рода и мощности'''
    prob_reject = n_rejects / n_iter
    left_bound, right_bound = proportion_confint(n_rejects, n_iter, method='wilson')
    return prob_reject, left_bound, right_bound

def visualization(prob_reject, left_bound, right_bound, show_pvals=False, pvals=None, alpha=0.05, figsize=(7, 2), title=None):
    '''Отрисовка интервала для вероятности отвержения критерия и гистограммы p-value (опционально)'''
    if show_pvals:  # построение гистограммы p-value (опционально)
        with sns.axes_style("whitegrid"):
            plt.figure(figsize=figsize)
            plt.subplot(1, 2, 1)
            plt.hist(pvals, bins=np.linspace(0, 1, 21), alpha=0.7, weights=np.ones(len(pvals)) / len(pvals))
            plt.title('Распределение p-value')
    with sns.axes_style("whitegrid"):  # отрисовка интервала для вероятности отвержения критерия
        if show_pvals:
            plt.subplot(1, 2, 2)
        else:
            plt.figure(figsize=figsize)
        plt.hlines(0, 0, 1, color='black', lw=2, alpha=0.6)
        plt.vlines(alpha, -1, 1, color='red', lw=5, linestyle='--', alpha=0.6)
        plt.fill_between([left_bound, right_bound], [0.15]*2, [-0.15]*2, color='green', alpha=0.6)
        plt.scatter(prob_reject, 0, s=300, marker='*', color='red')
        plt.xlim((min(alpha, left_bound)-1e-3, max(alpha, right_bound)+1e-3))
        plt.title(f'Доля отвержений = {100*prob_reject:.2f}%, интервал ({100*left_bound:.2f}%, {100*right_bound:.2f}%)')
        plt.suptitle(title)
        plt.ylim((-0.5, 0.5))
        plt.yticks([])
        plt.tight_layout()
        plt.show()

def draw_power(powers, left_powers, right_powers, effects_list=np.linspace(0, 0.1, 11), label=None, title='Графики мощности', new_figure=False):
    '''Построение графика мощности критерия'''
    if new_figure:
        plt.figure(figsize=(10, 4))
    plt.plot(effects_list, powers, label=label, lw=3)
    plt.fill_between(effects_list, left_powers, right_powers, alpha=0.3)
    if new_figure:
        plt.hlines(0.8, effects_list[0], effects_list[-1], color='black', alpha=0.5, label='Мощность 0.8')
    plt.legend()
    plt.xlabel('Размер относительного эффекта')
    plt.ylabel('Мощность')
    plt.title(title)

def add_effect(x, effect, target='pilot', relative_effect=True):
    '''Искусственное добавление эффекта в тестовую выборку'''
    x_copy = x.copy()
    x_copy[target] = x_copy[target]*(1+effect) if relative_effect else x_copy[target]+effect
    return x_copy

def run_experiments(test, generate_samples, n_iter=10000, sample_size=1000, effect=0, relative_effect=True, alpha=0.05, target='pilot', draw=False, title=None, show_pvals=True):
    '''Проведение серии AA-тестов на искусственных выборках'''
    n_rejects, pvals = 0, []  # количество отвержений H_0, значения p-value
    for _ in tqdm(range(n_iter), leave=False):  # Проведение экспериментов
        x_data, y_data = generate_samples(sample_size=sample_size)  # Генерируем выборки
        x_data = add_effect(x_data, effect, target=target, relative_effect=relative_effect)  # Добавляем эффект, если хотим оценить мощность
        pvals.append(test(x_data, y_data)[1])  # Применяем критерий
        n_rejects += (pvals[-1] < alpha)
    prob_reject, left_bound, right_bound = estimate_reject_prob(n_rejects, n_iter)  # Оценка вероятности отвержения
    if draw:  # Визуализация
        figsize = (14, 3) if show_pvals else (5, 2)
        visualization(prob_reject, left_bound, right_bound, show_pvals, pvals, alpha, figsize, title)
    return prob_reject, (left_bound, right_bound)

def estimate_power(test, generate_samples, n_iter=10000, sample_size=1000, effects_list=np.linspace(0, 0.1, 11), relative_effect=True, alpha=0.05, target='pilot'):
    '''Проведение серии тестов на искусственных выборках для оценки мощности'''
    powers, left_bounds, right_bounds = np.zeros(len(effects_list)), np.zeros(len(effects_list)), np.zeros(len(effects_list))
    for i, effect in tqdm(enumerate(effects_list), leave=False):
        powers[i], (left_bounds[i], right_bounds[i]) = run_experiments(test, generate_samples, n_iter=n_iter, sample_size=sample_size, effect=effect, relative_effect=relative_effect, alpha=alpha, target=target, draw=False)
    return powers, left_bounds, right_bounds
```

```python
def standard_2_step(x_data, y_data, target='pilot', covariate='prepilot'):
    X, y = y_data[[covariate]]-x_data[[covariate]], y_data[[target]]-x_data[[target]]
    th = LinearRegression(fit_intercept=True).fit(X,y).coef_[0][0]  #,cov_type='HC3'
    res = sm.OLS(
        endog=pd.concat([x_data[target]-th*x_data[covariate], y_data[target]-th*y_data[covariate]]), 
        exog=sm.add_constant(np.concatenate([np.zeros(len(x_data)), np.ones(len(y_data))]))
    ).fit(cov_type='HC3')
    return res.tvalues.iloc[1].item(), res.pvalues.iloc[1].item()

def combo_1_step(x_data, y_data, target='pilot', covariate='prepilot'):
    res = sm.OLS(
        endog=pd.concat([y_data[[target]],x_data[[target]]]).reset_index(drop=True),  # y
        exog=sm.add_constant(pd.concat([
            pd.DataFrame(np.concatenate([np.zeros(len(x_data)), np.ones(len(y_data))])),  # t
            pd.concat([y_data[[covariate]], x_data[[covariate]]]).reset_index(drop=True)  # X
        ], axis=1))
    ).fit(cov_type='HC3')
    return res.tvalues.iloc[1].item(), res.pvalues.iloc[1].item()

def combo_2_step(x_data, y_data, target='pilot', covariate='prepilot'):
    t = pd.DataFrame(np.concatenate([np.zeros(len(x_data)), np.ones(len(y_data))]), columns=[''])
    X, y = pd.concat([y_data[covariate],x_data[covariate]]).reset_index(drop=True), pd.concat([y_data[target],x_data[target]]).reset_index(drop=True)
    th = LinearRegression().fit(pd.concat([t,X],axis=1), y).coef_[-1]  #, cov_type='HC3'
    res = sm.OLS(endog=y-th*X, exog=sm.add_constant(t)).fit(cov_type='HC3')
    return res.tvalues.iloc[1].item(), res.pvalues.iloc[1].item()

tests = {
    'Без CUPED': lambda x, y: sps.ttest_ind(x['pilot'], y['pilot']),
    'Станд. 2-шаговый': standard_2_step,
    'Комб. 1-шаговый': combo_1_step,
    'Комб. 2-шаговый': combo_2_step
}
```

```python
for title, test in tests.items():
    run_experiments(test=test, generate_samples=generate_two_independent_samples, sample_size=200, n_iter=5000, draw=True, title=title)
```

*Все методы прошли валидацию*

```python
for i, (title, test) in enumerate(tests.items()):
    powers, Ls, Rs = estimate_power(test=test, generate_samples=generate_two_independent_samples, sample_size=200, n_iter=5000)
    draw_power(powers, Ls, Rs, label=title, new_figure=(i==0))
```

*Видно, что с CUPED мощность выше. При этом все методы с CUPED показывают близкие результаты*

```python
for title, test in tests.items():
    run_experiments(test=test, generate_samples=lambda **kwargs:generate_two_independent_samples(complex=True,**kwargs), sample_size=200, n_iter=5000, draw=True, title=title)
```

*На данных посложнее получилось, что не валидны все методы, кроме комб. 1-шагового*

```python
for i, (title, test) in enumerate(tests.items()):
    powers, Ls, Rs = estimate_power(test=test, generate_samples=lambda **kwargs:generate_two_independent_samples(complex=True,**kwargs), sample_size=200, n_iter=5000)
    draw_power(powers, Ls, Rs, label=title, new_figure=(i==0))
```
