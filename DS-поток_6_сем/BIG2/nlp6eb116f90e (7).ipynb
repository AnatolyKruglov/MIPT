Тональность по отношению к сущности в новостном тексте может происходить по крайней мере из трех разных источников:
* мнения автора текста;

````python
train_df, val_df = [pd.read_csv(f'/kaggle/input/rusentne/{x}.csv', sep='\t', usecols=['sentence','entity_tag','label']) for x in ['train_data','validation_data_labeled']]
for df in [train_df, val_df]:
    display(df.tail(2))  # тут лейблы -1, 0, 1
````

````python
stop_words, punc = nltk.corpus.stopwords.words("russian"), list(punctuation)
tokenizer=AutoTokenizer.from_pretrained("sberbank-ai/ruRoberta-large")
stemmer=SnowballStemmer('russian')

class RuSentNEDataset(Dataset):
    def __init__(self, df, stemmer=None):
        self.df = df
        self.df['label'] += 1  # иначе ловил ошибку, что таргет вне рэндж. Тут и далее лейблы 0, 1, 2
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        return (row['label'], row['sentence'].lower(), row['entity_tag'].lower())

def preprocess(dataset, stem=False):
    input_ids, attn_masks, labels = [], [], []
    for label, sentence, entity_tag in dataset:
        sentence = ' '.join([w for w in sentence.split() if w not in stop_words and w not in punc])  # уберем стоп-слова и пунктуацию
        if stem:  # стемминг
            sentence = ' '.join([stemmer.stem(w) for w in sentence.split()])
        enc = tokenizer.encode_plus(sentence, add_special_tokens=True, truncation=True, max_length=128, padding='max_length', \
                                    return_attention_mask=True, return_tensors='pt')
        input_ids.append(enc['input_ids'])
        attn_masks.append(enc['attention_mask'])
        labels.append(label)
    return D.from_dict({'input_ids': torch.cat(input_ids, dim=0), 'attention_mask': torch.cat(attn_masks, dim=0), 'labels': torch.tensor(labels)})

def train(name, train_dataset, val_dataset, num_epochs=8): 
    model = AutoModelForSequenceClassification.from_pretrained(name, num_labels=3)
    trainer = Trainer(
        model=model,
        args=TrainingArguments(output_dir="/kaggle/working", num_train_epochs=num_epochs, warmup_steps=500, learning_rate=5e-5,
            per_device_train_batch_size=16, per_device_eval_batch_size=16, logging_strategy="epoch", eval_strategy="epoch", save_strategy='no', #"epoch", 
            metric_for_best_model="f1", greater_is_better=True, report_to="none", #load_best_model_at_end=True, 
        ),
        train_dataset=train_dataset, eval_dataset=val_dataset,
        compute_metrics=lambda pred: {
            "f1": f1_score(pred.label_ids, pred.predictions.argmax(-1), average="macro", labels=[0,2]),
            "precision": precision_score(pred.label_ids, pred.predictions.argmax(-1), average="macro", labels=[0,2]),
            "recall": recall_score(pred.label_ids, pred.predictions.argmax(-1), average="macro", labels=[0,2])
        }  # labels=[0,2], чтобы только + и -
    )
    trainer.train()
    del model
````

````python
train_dataset, val_dataset = preprocess(RuSentNEDataset(train_df)), preprocess(RuSentNEDataset(val_df))
````

````python
train("sberbank-ai/ruRoberta-large", train_dataset, val_dataset)
````

````python
train("ai-forever/ruBert-base", train_dataset, val_dataset)
````

*Обучим модель*

````python
stop_words, punc = nltk.corpus.stopwords.words("russian"), list(punctuation)
tokenizer=AutoTokenizer.from_pretrained("sberbank-ai/ruRoberta-large")
stemmer=SnowballStemmer('russian')

class NsuDataset(Dataset):
    def __init__(self, df, stemmer=None):
        self.df = df
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        return (row['label'], row['txt'].lower())

def preprocess(dataset, stem=False):
    input_ids, attn_masks, labels = [], [], []
    for label, txt in dataset:
        txt = ' '.join([w for w in txt.split() if w not in stop_words and w not in punc])  # уберем стоп-слова и пунктуацию
        if stem:  # стемминг
            txt = ' '.join([stemmer.stem(w) for w in txt.split()])
        enc = tokenizer.encode_plus(txt, add_special_tokens=True, truncation=True, max_length=96, padding='max_length', \
                                    return_attention_mask=True, return_tensors='pt')
        input_ids.append(enc['input_ids'])
        attn_masks.append(enc['attention_mask'])
        labels.append(label)
    return D.from_dict({'input_ids': torch.cat(input_ids, dim=0), 'attention_mask': torch.cat(attn_masks, dim=0), 'labels': torch.tensor(labels)})

def train(name, train_dataset, val_dataset, num_epochs=8): 
    model = AutoModelForSequenceClassification.from_pretrained(name, num_labels=142)
    trainer = Trainer(
        model=model,
        args=TrainingArguments(output_dir="/kaggle/working", num_train_epochs=num_epochs, warmup_steps=500, learning_rate=5e-5,
            per_device_train_batch_size=16, per_device_eval_batch_size=16, logging_strategy="epoch", eval_strategy="epoch", save_strategy='no', #"epoch", 
            metric_for_best_model="accuracy", greater_is_better=True, report_to="none", #load_best_model_at_end=True, 
        ),
        train_dataset=train_dataset, eval_dataset=val_dataset,
        compute_metrics=lambda pred: {
            "accuracy": accuracy_score(pred.label_ids, pred.predictions.argmax(-1)),
        }
    )
    trainer.train()
````

````python
train_df, test_df = _load('train'), _load('test')
label_encoder = LabelEncoder()
train_df['label'] = label_encoder.fit_transform(train_df['label'])
test_df['label'] = label_encoder.transform(test_df['label'])
train_dataset, val_dataset = preprocess(NsuDataset(train_df)), preprocess(NsuDataset(test_df)) 
````

````python
train("ai-forever/ruBert-base", train_dataset, val_dataset)
````

С продуктовой точки зрения как раз и достаточно работать хорошо в среднем, преимущественно на самых частых классах, чтобы как можно больше юзеров были довольны ботом. Т.е. метрика все же отражает нужный нам аспект реальности*
