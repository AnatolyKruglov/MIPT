**1)** Постройте несколько графиков сопряженного распределения для разных параметров и охарактеризуйте, как его значения параметров соотносятся с априорными знаниями о монете.
Можно рассмотреть следующие случаи.
* Монета скорее честная. Тогда наиболее вероятны значения $p$ в окрестности $0.5$.
* Монета скорее нечестная, перевес неизвестен. Тогда наименее вероятны значения $p$ в окрестности $0.5$.
* Монета скорее нечестная, перевес в сторону герба. Тогда наиболее вероятны значения $p$ в окрестности $1$.
* Монета скорее честная, либо с небольшим перекосом вправо. Тогда наиболее вероятны значения $p$ в окрестности $\approx 0.6$.
* Ничего не известно. Тогда все значния равновероятны.

Для каждого случая из перечисленных выше постройте график плотности сопряженного распределения (на одной фигуре).

```python
def draw(A, B, title, p=None, x=np.linspace(0,1,1000)):
    plt.figure(figsize=(15,5))
    for label,a,b in zip(['Скорее честная','Скорее нечестная, перевес неизвестен','Скорее нечестная, перевес в сторону герба',
                            'Скорее честная, либо с небольшим перекосом вправо','Ничего не известно'], A, B):
        plt.plot(x, sps.beta(a,b).pdf(x), label=label)
    if p is not None:
        plt.scatter(p, 0, marker="o", color='red', label='Истинное значение')
    plt.xlabel("p")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

A,B = np.array([[9,9], [0.7,0.7], [4,1], [9,7], [1,1]]).T  #параметрыы бета-распр.-я
draw(A, B, title='Априорная плотность сопр. распр.-я')
```

Скорее честная: $a=b \geq 1$  
Скорее нечестная, перевес неизвестен: $a=b \leq 1$  
Скорее нечестная, перевес в сторону герба: $a>1 \geq b$  
Скорее честная, либо с небольшим перекосом вправо: $a>b \geq 1$  
Ничего не известно: $a=b=1$  

Реализуйте функцию, вычисляющую параметры априорного распределения:

```python
def bern_posterior_params(sample, a, b):  # кажется, в формулировке опечатка (Находит параметры апостериорного распределения)
    return a+sample.sum(), b-sample.sum()+len(sample) 
```

### Задача 6

**1)** В схеме испытаний Бернулли из задачи 4 рассмотрим онлайн-подход: наблюдения поступают батчами, в качестве априорного используется предыдущее апостериорное. Сначала реализуйте функцию генерации батчей:

```python
def get_sample_bern(theta=0.5, batch_size=5):
    return sps.bernoulli(p=theta).rvs(size=batch_size)
```

Допишите функцию, имитирующую онлайн подход:

```python
def bern_posterior_params(batch_count, a_prior, b_prior, theta=0.5, batch_size=5):
    '''
    :param batch_count: количество батчей в выборке
    :param a_prior, b_prior: параметры априорного распределения (до получения выборки)
    :param theta: истинный параметр распределения бернулли
    :param batch_size: размер батча. Размер выборки в таком случае <batch_size> * <batch_count>
    :returns: набор апостериорных параметров, полученных после каждого нового батча
    '''
    params = [[a_prior, b_prior]]
    for _ in range(batch_count):
        batch = get_sample_bern(theta, batch_size)
        a,b = params[-1]
        params.append([a+batch.sum(), b-batch.sum()+batch_size])
    return np.array(params)
```

Ниже приведена реализация вспомогательной функции

```python
def draw_posteriori(distr_class, post_params, xlim=None, grid=np.linspace(0,1)):
    '''Рисует серию графиков апостериорных плотностей.
    :param grid: сетка для построения графика
    :param distr_class: класс распределений из scipy.stats
    :param post_params: параметры апостериорных распределений, shape=(размер выборки, кол-во параметров)
    '''
    size = len(post_params) - 1
    plt.figure(figsize=(12,6))
    for n in range(size+1):
        plt.plot(grid, distr_class(post_params[n]).pdf(grid) if np.isscalar(post_params[n]) else distr_class(*post_params[n]).pdf(grid), \
                 label=f'n={n}: {str([round(x,3) for x in post_params[n]])}', lw=2.5, color=(1-n/size, n/size, 0))
    plt.title('Апостериорное распределение в зависимости от номера батча')
    plt.legend(fontsize=12)
    plt.xlim(xlim)
    plt.show()
```

Рассмотрите произвольную монету. Подберите два априорных распределения так, что в одном случае априорные знания должны совпадать с реальностью, а во втором нет. Рассмотрите как изменяется апостериорное распределение в зависимости от получения нового батча.

```python
draw_posteriori(sps.beta, bern_posterior_params(10, a_prior=5, b_prior=5))  # совпадает с реальностью
```

**2)** Проведите аналогичное исследование для выборки из распределения $Exp(\theta)$

```python
def get_sample_exp(theta=1, batch_size=5):
    return sps.expon(scale=theta).rvs(size=batch_size)

def exp_posterior_params(batch_count, a_prior, b_prior, theta=1, batch_size=5):
    params = [[a_prior, b_prior]]
    for _ in range(batch_count):
        batch = get_sample_exp(theta, batch_size)
        a,b = params[-1]
        params.append([a+batch.sum(), b+batch_size])
    params = np.array(params)
    return np.array([params[:,1], np.zeros_like(params[:,1]), 1/params[:,0]]).T
```

### Задача 7

Предположим, что выборка из распределения $\mathcal{N}(\theta_1, \theta_2^{-1})$ приходит постепенно по 5 наблюдений за один раз. До получения первой партии из 5 наблюдений отсутствует какая-либо информация о значениях параметров $\theta = (\theta_1, \theta_2)$. После получения очередной партии наблюдений вычисляется апостериорное распределение параметров, используя в качестве априорного распределения предыдущее апостериорное распределение. Промоделируйте и визуализируйте такой эксперимент и опишите, как меняются знания о параметрах в процессе эксперимента.

```python
def get_sample_norm(loc=0, scale=1, batch_size=5):
    return sps.norm(loc=loc, scale=scale).rvs(size=batch_size)

def norm_posterior_params(batch_count, a_prior, b_prior, loc=0, scale=1, batch_size=5):  # буквально предыдущая задача с небольшой поправкой
    params = [[a_prior, b_prior]]
    for _ in range(batch_count):
        batch = get_sample_norm(loc, scale, batch_size)
        a,b = params[-1]
        tmp = (a*b + batch.sum()) / (b+batch_size)
        params.append([tmp, np.sqrt(((scale**2 + loc**2)*b + batch.sum()**2) / (b+batch_size) - tmp**2)])
    return np.array(params)

draw_posteriori(sps.norm, norm_posterior_params(10, a_prior=0, b_prior=1), grid=np.linspace(-5, 5))
```

*Снова сходимся: колокол центрируется и сужается*
